```python
# Define a task class to hold task information
class Task:
  def __init__(self, description):
    self.description = description
    self.completed = False

# Function to display the current to-do list
def display_tasks(tasks):
  if not tasks:
    print("There are no tasks in the list.")
  else:
    print("** To-Do List **")
    for i, task in enumerate(tasks, start=1):
      status = "Completed" if task.completed else "Pending"
      print(f"{i}. {status}: {task.description}")

# Function to add a new task
def add_task(tasks):
  description = input("Enter task description: ")
  tasks.append(Task(description))
  print("Task added successfully!")

# Function to mark a task as completed
def mark_completed(tasks):
  if not tasks:
    print("There are no tasks to mark as completed.")
    return
  display_tasks(tasks)
  while True:
    try:
      task_number = int(input("Enter the number of the task to mark as completed (or 0 to exit): "))
      if task_number == 0:
        return
      tasks[task_number - 1].completed = True
      print(f"Task {task_number} marked as completed!")
      break
    except (IndexError, ValueError):
      print("Invalid task number. Please try again.")

# Function to display the main menu
def main_menu():
  print("\n** To-Do List App **")
  print("1. View Tasks")
  print("2. Add Task")
  print("3. Mark Task Completed")
  print("4. Exit")

def main():
  tasks = []  # List to store tasks

  while True:
    main_menu()
    choice = input("Enter your choice (1-4): ")

    if choice == '1':
      display_tasks(tasks.copy())  # Display a copy to avoid modifying completed status
    elif choice == '2':
      add_task(tasks)
    elif choice == '3':
      mark_completed(tasks)
    elif choice == '4':
      print("Exiting To-Do List App.")
      break
    else:
      print("Invalid choice. Please try again.")

if __name__ == "__main__":
  main()
```
